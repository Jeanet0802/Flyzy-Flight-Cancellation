```python
import pandas as pd 
import numpy as np
#load the dataset
data = pd.read_csv('Flyzy Flight Cancellation - Sheet1.csv')
data.iloc[1:4, 2:3]= np.NaN
data.iloc[1:4, 3:4]= "NA"
data.iloc[1:4, 4:5]= ""
data["None_col"]= None
data.head()
```

```python
null= pd.isnull(data)
null.head()
```

```python
pd.isnull(data).sum().sum()
```

```python
#Detect a list of missing values with df.isin()
missing_vals= ['NA', "",None, np.NaN]
missing= data.isin(missing_vals)
missing.head()
```

```python
data.fillna(0).head()
```

```python
# fill a list of missing values with a given value:
missing_vals= ['NA',"",None, np.NaN]
missing= data.isin(missing_vals)
data.mask(missing,"missing").head()
```

```python
# Check for missing values in each column again
missing_values = data.isnull().sum()

# Print the results
print("Missing values per column (before handling):")
print(missing_values)
```

```python
# Drop the 'Non_col' column
data = data.drop('None_col', axis=1)
```

```python
# Impute missing flight distances with mean
data['Flight_Distance'].fillna(data['Flight_Distance'].mean(), inplace=True)
```

```python
# Check for missing values in each column again
missing_values = data.isnull().sum()

# Print the results
print("Missing values per column (after handling):")
print(missing_values)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# Example: Plot 'Flight_Distance' on x-axis and 'Weather_Score' on y-axis
data.plot(kind='scatter', x='Flight_Distance', y='Weather_Score')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd


# Choose columns to analyze
columns_to_check = ['Flight_Distance', 'Weather_Score']

# Create a figure with a larger size
plt.figure(figsize=(10, 6))  # Adjust the width and height as needed

# Create boxplots
data[columns_to_check].plot(kind='box')

# Improve plot readability (optional)
plt.title('Boxplots for Outlier Detection')  # Add a title
plt.xlabel('Features')  # Add labels for axes
plt.ylabel('Values')
plt.grid(True)  # Add gridlines for better visualization

# Display the plot
plt.show()
```

Flight distance: There appear to be a few outliers on the higher end of the distribution for flight distance. This suggests there might be some flights in the dataset that are considerably longer than the majority of flights.

Weather score: There could potentially be a few outliers on both the lower and higher ends, indicating flights with exceptionally good or bad weather conditions compared to most flights.
